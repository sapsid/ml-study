```
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
SAMPLES=500
```

```
data = np.random.uniform(-50,50,[SAMPLES,])
```

```
b = tf.keras.activations.linear(data)
plt.plot(data, b)
plt.show()
```

```
b = tf.keras.activations.relu(data)
plt.scatter(data, b.numpy())
plt.show()
```

```
b = tf.keras.activations.sigmoid(data)
plt.scatter(data, b.numpy())
plt.show()
```

```
b = tf.keras.activations.tanh(data)
plt.scatter(data, b.numpy())
plt.show()
```

```
data = tf.random.normal([10, 10])
b = tf.keras.activations.softmax(data)
plt.scatter(data, b.numpy())
plt.show()
```
